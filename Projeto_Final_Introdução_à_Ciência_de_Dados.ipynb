**10.** Teste as seguintes hipóteses a respeito dos dados. Em todos os testes, assuma quaisquer premissas necessárias (ex.: normalidade, homocedasticidade,...)

*a)* Pinguins femininos têm uma média de comprimento de bico maior que os pinguins masculinos.

*b)* A ilha no qual o animal se encontra tem efeito estatisticamente significante na massa média.

**Observações:** Para resolver esse exercício apresente:

- as hipóteses formalmente enunciadas
- o método aplicado para testá-las
- o nível de significância adotado
- o valor da estatística de teste e o p-valor
- a tomada de decisão justificada

`````python
#modulos python
import warnings
import  numpy  as  np 
import  pandas  as  pd 
import  scipy.stats  as  sts 
import  statsmodels.api  as  sm 
import  matplotlib.pyplot  as  plt 
import  seaborn  as  sns

# ignorar warnings 
warnings.filterwarnings('ignore')

# carregando dados
data = sns.load_dataset('penguins')

print('Hipótese: a) Pinguins femininos têm uma média de comprimento de bico maior que os pinguins masculinos. ' )

# selecionar subgrupos para realizar os testes estatísticos

femeas = data.loc[data['sex'] == 'Female', 'bill_length_mm']
machos = data.loc[data['sex'] == 'Male', 'bill_length_mm']


# calculando a estatística de teste f
f = np.var(femeas, ddof=1) / np.var(machos, ddof = 1)
num = femeas.shape[0] - 1
den = machos.shape[0] - 1
p_value = 1 - sts.f.cdf(f, num, den)

print(f'Estatística de Teste F: {f}')
print(f'p_valor: {p_value}')

# aplicando teste estatístico t de Student - variâncias diferentes
print('Resultado do teste t: ', sts.ttest_ind(femeas, machos, equal_var = False))

simul1 = np.random.normal(loc = femeas.mean(), scale = femeas.std()/np.sqrt(femeas.shape[0]), size = 10000)
simul2 = np.random.normal(loc = machos.mean(), scale = machos.std()/np.sqrt(machos.shape[0]), size = 10000)

plt.figure(figsize=(8,8))
sns.distplot(simul1, color = 'skyblue', kde = True, label = 'Fêmeas')
sns.distplot(simul2, color = 'lightgreen', kde = True, label = 'Machos')
plt.axvline(x = femeas.mean(), color = 'blue', ls = '--')
plt.axvline(x = machos.mean(), color = 'green', ls = '--')
plt.xlabel('Comprimento do bico (mm)', size = 14)
plt.ylabel('Densidade de Probabilidade', size = 14)
plt.title('Comparação do Comprimento do bico', size = 16)
plt.legend(loc = 'best', prop = {'size': 14})
plt.show()


print('Hipótese: b) A ilha no qual o animal se encontra tem efeito estatisticamente significante na massa média.')

# selecionar subgrupos para realizar os testes estatísticos

ilha1 = data.loc[data['island'] == 'Torgersen', 'body_mass_g']
ilha2 = data.loc[data['island'] == 'Biscoe', 'body_mass_g']
ilha3 = data.loc[data['island'] == 'Dream', 'body_mass_g']

# aplicando análise de variância
print(sts.f_oneway(ilha1, ilha2, ilha3))

simul4 = np.random.normal(loc = ilha1.mean(), scale = ilha1.std()/np.sqrt(ilha1.shape[0]), size = 10000)
simul5 = np.random.normal(loc = ilha2.mean(), scale = ilha2.std()/np.sqrt(ilha2.shape[0]), size = 10000)
simul6 = np.random.normal(loc = ilha3.mean(), scale = ilha3.std()/np.sqrt(ilha3.shape[0]), size = 10000)


plt.figure(figsize=(8,8))
sns.distplot(simul4, color = 'skyblue', kde = True, label = 'Ilha 1')
sns.distplot(simul5, color = 'lightgreen', kde = True, label = 'Ilha 2')
sns.distplot(simul6, color = 'pink', kde = True, label = 'Ilha 3')

plt.axvline(x = ilha1.mean(), color = 'blue', ls = '--')
plt.axvline(x = ilha2.mean(), color = 'green', ls = '--')
plt.axvline(x = ilha3.mean(), color = 'red', ls = '--')

plt.xlabel('Massa corporal (g)', size = 14)
plt.ylabel('Densidade de Probabilidade', size = 14)
plt.title('Comparação da massa corporal', size = 16)
plt.legend(loc = 'best', prop = {'size': 14})
plt.show()

print('Dá para se observar que os machos tem bicos mais compridos que as fêmeas e que os moradores da ilha Biscoe diferem dos demais, possuindo mais massa corporal.')



`````
